### Dataset3

```python
d3 = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv')
```

```python
d3.head()
```

```python
d3.info()
```

```python
d3['Year'] = pd.to_datetime(d3.Year,format='%Y')
d3.info()
```

```python
d3.head()
```

```python
d3 = d3.set_index('Year',drop=True)
d3
```

```python
#Group the year by decades and sum the values
```

```python
# Uses resample to sum each decade
d3s = d3.resample('10AS').sum()
d3s
```

```python
# Uses resample to get the max value only for the "Population" column
population = d3['Population'].resample('10AS').max()
population
```

```python
# Updating the "Population" column
d3s['Population'] = population
d3s
```

```python
d3.loc[d3.index[[3,4,8]],['Murder','Robbery']]
```

```python
d3[d3['Murder']>24000]
```

```python
d3[d3['Murder'].between(20000,24000)]
```

```python
d3.groupby('Year')['Murder'].mean()
```

```python
d3.sort_values(['Murder','Violent'],ascending=[False,True])
```
